```python
import pandas as pd
```

```python
# Read in the green taxi data
df_green_taxi = pd.read_parquet("data/green_tripdata_2015-01.parquet", engine="pyarrow")
df_green_taxi.head()
```

```python
df_green_taxi['pickup_date'] = df_green_taxi['lpep_pickup_datetime'].dt.date
df_green_taxi['dropoff_date'] = df_green_taxi['lpep_dropoff_datetime'].dt.date
```

```python
df_green_taxi
```

```python
df_green_taxi['year'] = df_green_taxi['lpep_pickup_datetime'].dt.year
df_green_taxi.to_csv("data/green_tripdata_2015-01_modified.csv", index=False)
```
